```python
import tensorflow as tf
```

```python
import numpy as np
from tensorflow import keras
```

## Define and compile the NN

Simplest Neural Network:
* one layer
*  one neuron 
* input shape one value

```python
model = tf.keras.Sequential(
    [keras.layers.Dense(units=1, input_shape=[1])])
```

Compile NN with `loss` and `optimizer`

`loss`      - measures guessed values agaisnt know correct values and measures the fit(how good or bad)
 
`optimizer` - used to make another guess based off the loss function result with the goal of shrinking the loss.

`mean_squared_error` - 

`sgd` : stochastic gradient descent 

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

## Provide Data

```python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)
```

```python
model.fit(xs,ys, epochs=500)
```

## Use Model

```python
print(model.predict([10.0]))
```
